```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from regression import *

df = pd.read_csv("team20/dataset2/Test-50.csv")
df.columns = df.columns.str.lower()
x1 = df.x1.to_numpy()
x2 = df.x2.to_numpy()
y = df.output.to_numpy()

fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.set_title("Training Dataset 2b")
ax.scatter(x1, x2, y)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")
ax.set_proj_type("ortho")
```

```python
f = train_bivar(x1, x2, y, 9, 10)
```

```python
# Generate a grid of values over the input range
x1_grid, x2_grid = np.meshgrid(
    np.linspace(min(x1), max(x1)), np.linspace(min(x2), max(x2))
)
y_grid = np.array(
    [f(x1_val, x2_val) for x1_val, x2_val in zip(x1_grid.ravel(), x2_grid.ravel())]
)
y_grid = y_grid.reshape(x1_grid.shape)
ax.plot_wireframe(x1_grid, x2_grid, y_grid, color="g", linewidth=0.2)

fig
```
