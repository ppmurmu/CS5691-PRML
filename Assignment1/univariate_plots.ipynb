# PRML Assignment 01: Univariant Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from regression import *

df = pd.read_csv("./team20/dataset1/train_1b.csv")
df.columns = df.columns.str.lower()
x = df.x.to_numpy()
y = df.y.to_numpy()
```

```python
fig, ax = plt.subplots()
ax.set_title("Training Dataset 1b")
ax.set_xlabel("X_data")
ax.set_ylabel("Model output Yn")
ax.set_ylim(-0.7, 1.6)
ax.set_xlim(min(x) - 0.5, max(x) + 0.5)

data_plt = ax.scatter(df.x, df.y)
data_plt.set_label("Training data")
data_plt.set_facecolors("none")
data_plt.set_edgecolors("b")
```

```python
f = train_univar(x, y, 9)
x_space = np.linspace(min(x) - 0.5, max(x) + 0.5)
ax.plot(x_space, f(x_space), label="Curvefit without regularization")
ax.legend(loc="upper right")

fig
```

```python
f_reg = train_univar(x, y, 9, 10)
ax.plot(x_space, f_reg(x_space), label="Curvefit with regularization")
ax.legend(loc="upper right")

fig
```
