# **Testing dataset 1a**


```python
#requirements
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from regression import *
```

```python
#load data
df = pd.read_csv("./dataset1/test.csv")
df.columns = df.columns.str.lower()
x = df.x.to_numpy()
y = df.y.to_numpy()

fig, ax = plt.subplots()
ax.set_title("Training Dataset 1a")
ax.set_xlabel("X_data")
ax.set_ylabel("Model output Yn")
ax.set_ylim(-0.5, 1.6)
ax.set_xlim(min(x) - 0.5, max(x) + 0.5)

data_plt = ax.scatter(df.x, df.y)
data_plt.set_label("Training data")
data_plt.set_facecolors("none")
data_plt.set_edgecolors("b")

```

## for degree 3


```python
#without regularization for degree 3
f = train_univar(x, y, 3)
x_space = np.linspace(min(x) - 0.5, max(x) + 0.5)
line1, =ax.plot(x_space, f(x_space), label="Curvefit without regularization")
ax.legend(loc="upper right")
ax.set_title("Training Dataset 1a with degree 3")

#RMSE
c= RMSE(f(x), y)
print(c)


#fig.savefig("./train-1a-deg3-0.svg", format="svg", dpi=500)
```

## for degree 6


```python
#without regularization for degree 6
line1.remove();
f = train_univar(x, y, 6)
x_space = np.linspace(min(x) - 0.5, max(x) + 0.5)
line1, =ax.plot(x_space, f(x_space), label="Curvefit without regularization")
ax.legend(loc="upper right")
ax.set_title("Training Dataset 1a with degree 6")
c= RMSE(f(x), y)

#RMSE
c= RMSE(f(x), y)
print(c)
#fig.savefig("./train-1a-deg6-0.svg", format="svg", dpi=500)
```
