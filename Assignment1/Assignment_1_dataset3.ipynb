```python
#data set 1a
best_degree=9
best_lambda=0.001
df = pd.read_csv("./dataset1/train_1a.csv")
df.columns = df.columns.str.lower()
x = df.x.to_numpy()
y_train = df.y.to_numpy()


y_predict = train_univar(x, y_train, best_degree, best_lambda)


#Create the plot
fig=plt.figure(figsize=(10, 8))
plt.scatter(y_train, y_predict(x), color='blue', label='output')

# Add the diagonal line
min_val = min(y_train.min(), y_predict(x).min())
max_val = max(y_train.max(), y_predict(x).max())
plt.plot([min_val, max_val], [min_val, max_val], 'k-', lw=2)

# Set labels and title
plt.xlabel('Y train')
plt.ylabel('Y predicted train')
plt.title(f'Y train vs Y predicted_train for train-1a\nDegree = {best_degree}    ={best_lambda} ')

# Add legend
plt.legend()

# Set axis limits
plt.xlim(min_val, max_val)
plt.ylim(min_val, max_val)

# Display the plot
#plt.tight_layout()
plt.show()

fig.savefig("./scatter_plot_1a_train.svg", format="svg", dpi=500)
```
