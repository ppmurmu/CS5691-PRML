```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Custom Tanh function with slope parameter (beta)
class CustomTanh(nn.Module):
    def __init__(self, beta=1.0):
        super(CustomTanh, self).__init__()
        self.beta = beta

    def forward(self, x):
        return torch.tanh(self.beta * x)

class MLFFNNFunctionApproximation:
    def __init__(self, input_dim, hidden_dims, output_dim, activation_fn=CustomTanh(beta=1.0)):
        if len(hidden_dims) < 1 or hidden_dims[0] < 2:
            raise ValueError("Hidden layer must have at least 2 nodes for plotting surfaces.")

        layers = []
        self.hidden_nodes_outputs = {}  # To store intermediate outputs

        # Input layer
        layers.append(nn.Linear(input_dim, hidden_dims[0]))
        layers.append(activation_fn)

        # Hidden layers
        for i in range(1, len(hidden_dims)):
            layers.append(nn.Linear(hidden_dims[i - 1], hidden_dims[i]))
            layers.append(activation_fn)

        # Output layer (linear activation)
        layers.append(nn.Linear(hidden_dims[-1], output_dim))

        # Define the model
        self.model = nn.Sequential(*layers)

        # Loss and optimizer
        self.criterion = nn.MSELoss()
        self.optimizer = optim.SGD(self.model.parameters(), lr=0.01, momentum=0.9)

        # Scalers for input and output data
        self.input_scaler = MinMaxScaler(feature_range=(-1, 1))
        self.output_scaler = MinMaxScaler(feature_range=(-1, 1))

    def scale_data(self, X_train, y_train, X_test, y_test):
        # Scale input data
        X_train_scaled = self.input_scaler.fit_transform(X_train)
        X_test_scaled = self.input_scaler.transform(X_test)

        # Reshape and scale output data
        y_train = y_train.reshape(-1, 1)
        y_test = y_test.reshape(-1, 1)
        y_train_scaled = self.output_scaler.fit_transform(y_train)
        y_test_scaled = self.output_scaler.transform(y_test)

        return X_train_scaled, y_train_scaled, X_test_scaled, y_test_scaled

    def get_hidden_outputs(self, x):
        """
        Capture the outputs of the hidden layers.
        """
        with torch.no_grad():
            for layer in self.model:
                x = layer(x)
                if isinstance(layer, nn.Linear):
                    # Store output of each hidden layer
                    self.hidden_nodes_outputs[layer] = x.numpy()  # Save the output from the last hidden layer
        return x.numpy()  # Return the output from the final layer

    def train(self, X_train, y_train, epochs=1000, avg_error_threshold=0.001, plot_interval=10):
        loss_history = []
        inputs = torch.tensor(X_train, dtype=torch.float32)
        targets = torch.tensor(y_train, dtype=torch.float32)
        previous_avg_loss = None

        save_epochs = [1, 10, 50]  # Epochs to save hidden and output layer data
        tracked_outputs = {epoch: {'hidden': None, 'output': None} for epoch in save_epochs}
        convergence_epoch = None

        for epoch in range(epochs):
            outputs = self.model(inputs)
            loss = self.criterion(outputs, targets)

            self.optimizer.zero_grad()
            loss.backward()
            self.optimizer.step()

            current_loss = loss.item()
            loss_history.append(current_loss)

            # Capture outputs at specific epochs
            if epoch in save_epochs:
                hidden_layer_output = self.get_hidden_outputs(inputs)  # Get last hidden layer output
                if hidden_layer_output is not None:
                    tracked_outputs[epoch]['hidden'] = hidden_layer_output
                    tracked_outputs[epoch]['output'] = outputs.detach().numpy()

            # Compute average loss over all epochs so far
            avg_loss = np.mean(loss_history)  # Average over all epochs so far

            # Print progress every 10 epochs
            if epoch % plot_interval == 0:
                print(f'Epoch [{epoch}/{epochs}], Loss: {current_loss:.6f}, Avg Loss: {avg_loss:.6f}')

            # Convergence criterion: Break if the change in the average loss is below the threshold
            if previous_avg_loss is not None and abs(avg_loss - previous_avg_loss) < avg_error_threshold:
                convergence_epoch = epoch
                print(f"Converged at epoch {epoch} with change in average loss: {abs(avg_loss - previous_avg_loss):.6f}")
                break

            previous_avg_loss = avg_loss  # Update previous average loss for next comparison

        # Save the final converged outputs
        if convergence_epoch:
            hidden_layer_output = self.get_hidden_outputs(inputs)
            tracked_outputs[convergence_epoch] = {'hidden': hidden_layer_output, 'output': outputs.detach().numpy()}

        return loss_history, tracked_outputs

    def predict(self, X_test):
        inputs = torch.tensor(X_test, dtype=torch.float32)
        with torch.no_grad():
            predictions = self.model(inputs)
        return self.output_scaler.inverse_transform(predictions.numpy())

    def plot_training_error(self, loss_history):
        plt.plot(loss_history)
        plt.xlabel('Epoch')
        plt.ylabel('Training Error')
        plt.title('Training Error vs Epoch')
        plt.show()

    def plot_predictions_vs_actual(self, y_test, y_pred, data):
        plt.scatter(range(len(y_test)), y_test, c='blue', label='Actual Values', alpha=0.7)
        plt.scatter(range(len(y_pred)), y_pred, c='orange', label='Predicted Values', alpha=0.7)

        plt.xlabel('Index')
        plt.ylabel('Values')
        plt.title('Predicted vs Actual Values ' + data)
        plt.legend()
        plt.grid(True)
        plt.show()

    def plot_surface(self, hidden_outputs, output_outputs, title, node_indices):
        """
        Plot surface for the outputs of two hidden layer nodes and the output node.
        Only works when input dimension is 2.
        :param hidden_outputs: The outputs of the hidden layer.
        :param output_outputs: The outputs of the output layer.
        :param title: Title for the plot.
        :param node_indices: A tuple with indices of two hidden layer nodes to plot.
        """
        # Check if the inputs are None
        if hidden_outputs is None or output_outputs is None:
            print("Error: One or more of the provided outputs is None. Skipping the plot.")
            return

        # Check if hidden outputs have at least 2 dimensions for plotting
        if hidden_outputs.shape[1] < 2:
            print(f"Error: Hidden outputs have {hidden_outputs.shape[1]} dimensions, which is not enough to plot. Skipping the plot.")
            return

        fig = plt.figure(figsize=(16, 8))

        # Prepare the grid for input features
        X1_range = np.linspace(-1, 1, 100)  # Based on scaling
        X2_range = np.linspace(-1, 1, 100)
        X1, X2 = np.meshgrid(X1_range, X2_range)

        try:
            # Reshape hidden outputs and output node outputs
            hidden_outputs_reshaped_1 = hidden_outputs[:, node_indices[0]].reshape(X1.shape)
            hidden_outputs_reshaped_2 = hidden_outputs[:, node_indices[1]].reshape(X1.shape)
            output_outputs_reshaped = output_outputs.reshape(X1.shape)

            # Surface for the first selected hidden node
            ax1 = fig.add_subplot(131, projection='3d')
            ax1.plot_surface(X1, X2, hidden_outputs_reshaped_1, cmap='viridis', edgecolor='none')
            ax1.set_title(f'Hidden Node {node_indices[0]}')
            ax1.set_xlabel('Input Feature 1')
            ax1.set_ylabel('Input Feature 2')
            ax1.set_zlabel('Node Output')

            # Surface for the second selected hidden node
            ax2 = fig.add_subplot(132, projection='3d')
            ax2.plot_surface(X1, X2, hidden_outputs_reshaped_2, cmap='plasma', edgecolor='none')
            ax2.set_title(f'Hidden Node {node_indices[1]}')
            ax2.set_xlabel('Input Feature 1')
            ax2.set_ylabel('Input Feature 2')
            ax2.set_zlabel('Node Output')

            # Surface for output node
            ax3 = fig.add_subplot(133, projection='3d')
            ax3.plot_surface(X1, X2, output_outputs_reshaped, cmap='inferno', edgecolor='none')
            ax3.set_title('Output Node')
            ax3.set_xlabel('Input Feature 1')
            ax3.set_ylabel('Input Feature 2')
            ax3.set_zlabel('Node Output')

            plt.suptitle(title)
            plt.show()

        except ValueError as e:
            print(f"Error during reshaping or plotting: {e}")


# Example usage (continuing from before)
input_dim = 2
hidden_dims = [8]
output_dim = 1

mlffnn = MLFFNNFunctionApproximation(input_dim, hidden_dims, output_dim)

# Example training and testing data (replace with your actual data)
X_train = x1_train
y_train = y1_train  # Replace with your data
X_test = x1_test
y_test = y1_test  # Replace with your data

# Scale data
X_train_scaled, y_train_scaled, X_test_scaled, y_test_scaled = mlffnn.scale_data(X_train, y_train, X_test, y_test)

# Train the model
loss_history, tracked_outputs = mlffnn.train(X_train_scaled, y_train_scaled, epochs=1000)

# Plot training error
mlffnn.plot_training_error(loss_history)

# Make predictions
y_pred = mlffnn.predict(X_test_scaled)

# Plot predictions vs actual
mlffnn.plot_predictions_vs_actual(y_test, y_pred, 'Test Data')

# Plot surfaces for selected hidden and output nodes
for epoch in [1, 10, 50]:
    hidden_out = tracked_outputs[epoch]['hidden']
    output_out = tracked_outputs[epoch]['output']
    mlffnn.plot_surface(hidden_out, output_out, f'Surface Plot at Epoch {epoch}', node_indices=(0, 1))


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from matplotlib.colors import ListedColormap
import seaborn as sns

class GMMClassifier:
    def __init__(self, n_components=8):
        self.n_components = n_components  # Number of Gaussians per class
        self.gmm_full = None              # GMM for full covariance
        self.gmm_diag = None               # GMM for diagonal covariance

    def fit(self, X_train, y_train):
        """Fit GMM classifiers for full and diagonal covariance."""
        self.gmm_full = self._fit_gmm_classifier(X_train, y_train, covariance_type='full')
        self.gmm_diag = self._fit_gmm_classifier(X_train, y_train, covariance_type='diag')

    def _fit_gmm_classifier(self, X_train, y_train, covariance_type='full'):
        """Fit a GMM for each class and return the classifiers."""
        gmm_classifiers = []

        # Fit GMM for each class
        for class_label in np.unique(y_train):
            X_class = X_train[y_train == class_label]
            gmm = GaussianMixture(n_components=self.n_components, covariance_type=covariance_type, random_state=42)
            gmm.fit(X_class)
            gmm_classifiers.append(gmm)

        return gmm_classifiers

    def predict(self, X):
        """Predict class labels based on GMMs."""
        predictions_full = self._predict_gmm(X, self.gmm_full)
        predictions_diag = self._predict_gmm(X, self.gmm_diag)
        return predictions_full, predictions_diag

    def _predict_gmm(self, X, gmm_classifiers):
        """Compute likelihoods and return predicted class labels."""
        likelihoods = np.array([gmm.score_samples(X) for gmm in gmm_classifiers]).T
        return np.argmax(likelihoods, axis=1)

    def evaluate(self, X_test, y_test):
        """Evaluate the accuracy of both classifiers."""
        y_pred_full, y_pred_diag = self.predict(X_test)
        accuracy_full = accuracy_score(y_test, y_pred_full)
        accuracy_diag = accuracy_score(y_test, y_pred_diag)
        return accuracy_full, accuracy_diag

    def plot_decision_regions(self, X, y, covariance_type='full'):
        """Plot decision regions based on GMM classifiers."""
        if covariance_type not in ['full', 'diag']:
            raise ValueError("covariance_type must be either 'full' or 'diag'.")

        gmm_classifiers = self.gmm_full if covariance_type == 'full' else self.gmm_diag

        # Define color maps
        cmap_light = ListedColormap(['#FFDDDD', '#DDDDFF'])  # Light colors for regions
        cmap_bold = ListedColormap(['#FF0000', '#00FF00'])    # Bright colors for class points

        # Generate meshgrid for plotting
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

        # Initialize Z for class predictions
        Z = np.zeros(xx.shape)

        # Plot individual Gaussian components and decision regions
        plt.figure(figsize=(10, 8))

        # Loop through each Gaussian component
        for gmm in gmm_classifiers:
            # Compute the probability density function for each Gaussian
            pdf = np.exp(gmm.score_samples(np.c_[xx.ravel(), yy.ravel()]))
            pdf = pdf.reshape(xx.shape)

            # Plot the contour lines for this Gaussian component using a dark color
            plt.contour(xx, yy, pdf, levels=5, cmap='Blues', alpha=0.99, linewidths=2)  # Dark grey for contours

        # Predict class labels for each point in the meshgrid
        Z = self._predict_gmm(np.c_[xx.ravel(), yy.ravel()], gmm_classifiers)
        Z = Z.reshape(xx.shape)

        # Plot overall decision regions
        plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.5)

        # Scatter plot for the training points
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=40)

        plt.title(f'GMM Classifier Decision Regions ({covariance_type.capitalize()} Covariance)')
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())
        plt.grid(True)  # Add grid for better visualization

        # Create a legend for the classes
        handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=cmap_bold.colors[i], markersize=10)
                   for i in range(len(np.unique(y)))]
        plt.legend(handles, [f'Class {i}' for i in np.unique(y)], title="Classes")

        plt.show()


    def plot_confusion_matrix(self,y_true, y_pred, class_names):
        """
        Compute and plot the confusion matrix.

        Parameters:
        - y_true: true class labels
        - y_pred: predicted class labels
        - class_names: list of class names for labeling the axes
        """
        # Compute confusion matrix
        cm = confusion_matrix(y_true, y_pred)

        # Create a heatmap for the confusion matrix
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=class_names, yticklabels=class_names)
        plt.title('Confusion Matrix')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.show()


# Example usage with actual dataset
# Replace x3_train, x3_test, y3_train, y3_test with your dataset
X_train, X_test, y_train, y_test = x3_train, x3_test, y3_train, y3_test

# Create an instance of GMMClassifier
gmm_classifier = GMMClassifier(n_components=16) #2*8 2 classes 8 components per class

# Fit the classifier
gmm_classifier.fit(X_train, y_train)

# Evaluate the classifier for training data
accuracy_full, accuracy_diag = gmm_classifier.evaluate(X_train, y_train)
print(f"Accuracy for training data(Full Covariance): {accuracy_full * 100:.2f}%")
print(f"Accuracy for training data(Diagonal Covariance): {accuracy_diag * 100:.2f}%")

#plot confusion matrix
# Assuming you have the predicted labels
y_pred_full, y_pred_diag = gmm_classifier.predict(X_train)

# Plot confusion matrix for Full Covariance GMM
print("Confusion Matrix for training data with Full Covariance GMM:")
gmm_classifier.plot_confusion_matrix(y_train, y_pred_full, class_names=["Class 0", "Class 1"])

# Plot confusion matrix for Diagonal Covariance GMM
print("Confusion Matrix for training data with Diagonal Covariance GMM:")
gmm_classifier.plot_confusion_matrix(y_train, y_pred_diag, class_names=["Class 0", "Class 1"])

# Evaluate the classifier for test data
accuracy_full, accuracy_diag = gmm_classifier.evaluate(X_test, y_test)
print(f"Accuracy for test data(Full Covariance): {accuracy_full * 100:.2f}%")
print(f"Accuracy for test data(Diagonal Covariance): {accuracy_diag * 100:.2f}%")

#plot confusion matrix
# Assuming you have the predicted labels
y_pred_full, y_pred_diag = gmm_classifier.predict(X_test)

# Plot confusion matrix for Full Covariance GMM
print("Confusion Matrix for test data with Full Covariance GMM:")
gmm_classifier.plot_confusion_matrix(y_test, y_pred_full, class_names=["Class 0", "Class 1"])

# Plot confusion matrix for Diagonal Covariance GMM
print("Confusion Matrix for test data with Diagonal Covariance GMM:")
gmm_classifier.plot_confusion_matrix(y_test, y_pred_diag, class_names=["Class 0", "Class 1"])


# Plot decision regions
gmm_classifier.plot_decision_regions(X_train, y_train, covariance_type='full')
gmm_classifier.plot_decision_regions(X_train, y_train, covariance_type='diag')




X_train, X_test, y_train, y_test = x4_train, x4_test, y4_train, y4_test
gmm_classifier = GMMClassifier(n_components=15) # 5 classes 3 components per class

# Fit the classifier
gmm_classifier.fit(X_train, y_train)

# Evaluate the classifier for training data
accuracy_full, accuracy_diag = gmm_classifier.evaluate(X_train, y_train)
print(f"Accuracy for training data(Full Covariance): {accuracy_full * 100:.2f}%")
print(f"Accuracy for training data(Diagonal Covariance): {accuracy_diag * 100:.2f}%")

#plot confusion matrix
# Assuming you have the predicted labels
y_pred_full, y_pred_diag = gmm_classifier.predict(X_train)

# Plot confusion matrix for Full Covariance GMM
print("Confusion Matrix for training data with Full Covariance GMM:")
gmm_classifier.plot_confusion_matrix(y_train, y_pred_full, class_names=["Class 0", "Class 1","Class 2", "Class 3","Class 4"])

# Plot confusion matrix for Diagonal Covariance GMM
print("Confusion Matrix for training data with Diagonal Covariance GMM:")
gmm_classifier.plot_confusion_matrix(y_train, y_pred_diag, class_names=["Class 0", "Class 1","Class 2", "Class 3","Class 4"])

# Evaluate the classifier for test data
accuracy_full, accuracy_diag = gmm_classifier.evaluate(X_test, y_test)
print(f"Accuracy for test data(Full Covariance): {accuracy_full * 100:.2f}%")
print(f"Accuracy for test data(Diagonal Covariance): {accuracy_diag * 100:.2f}%")

#plot confusion matrix
# Assuming you have the predicted labels
y_pred_full, y_pred_diag = gmm_classifier.predict(X_test)

# Plot confusion matrix for Full Covariance GMM
print("Confusion Matrix for test data with Full Covariance GMM:")
gmm_classifier.plot_confusion_matrix(y_test, y_pred_full, class_names=["Class 0", "Class 1","Class 2", "Class 3","Class 4"])

# Plot confusion matrix for Diagonal Covariance GMM
print("Confusion Matrix for test data with Diagonal Covariance GMM:")
gmm_classifier.plot_confusion_matrix(y_test, y_pred_diag, class_names=["Class 0", "Class 1","Class 2", "Class 3","Class 4"])
```

```python
class MLFFNNClassifier(nn.Module):
    def __init__(self, input_dim, hidden1_dim, hidden2_dim, output_dim):
        super(MLFFNNClassifier, self).__init__()
        self.hidden1 = nn.Linear(input_dim, hidden1_dim)
        self.hidden2 = nn.Linear(hidden1_dim, hidden2_dim)
        self.output = nn.Linear(hidden2_dim, output_dim)

    def forward(self, x):
        x = torch.tanh(self.hidden1(x))
        x = torch.tanh(self.hidden2(x))
        x = torch.softmax(self.output(x), dim=1)
        return x

class MLFFNNClassificationTrainer:
    def __init__(self, input_dim, hidden1_dim, hidden2_dim, output_dim):
        self.model = MLFFNNClassifier(input_dim, hidden1_dim, hidden2_dim, output_dim)
        self.criterion = nn.CrossEntropyLoss()
        self.optimizer = optim.SGD(self.model.parameters(), lr=0.7, momentum=0.9)

    def train(self, X_train, y_train, epochs=1000, threshold=0.001):
        loss_history = []
        for epoch in range(epochs):
            inputs = torch.tensor(X_train, dtype=torch.float32)
            targets = torch.tensor(y_train, dtype=torch.long)

            outputs = self.model(inputs)
            loss = self.criterion(outputs, targets)

            self.optimizer.zero_grad()
            loss.backward()
            self.optimizer.step()

            loss_history.append(loss.item())
            if epoch % 10 == 0:
                print(f'Epoch [{epoch}/{epochs}], Loss: {loss.item():.4f}')

            if loss.item() < threshold:
                break

        return loss_history

    def predict(self, X_test):
        inputs = torch.tensor(X_test, dtype=torch.float32)
        with torch.no_grad():
            outputs = self.model(inputs)
        return torch.argmax(outputs, dim=1).numpy()

    def plot_training_error(self, loss_history):
        plt.plot(loss_history)
        plt.xlabel('Epoch')
        plt.ylabel('Training Error')
        plt.title('Training Error vs Epoch')
        plt.show()

# Sample usage
mlffnn_classifier = MLFFNNClassificationTrainer(input_dim=2, hidden1_dim=12, hidden2_dim=8, output_dim=2)
loss_history = mlffnn_classifier.train(X_train, y_train)
mlffnn_classifier.plot_training_error(loss_history)
```

```python
def plot_decision_regions(model, data, labels, resolution=0.02):
    x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
    y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, resolution),
                         np.arange(y_min, y_max, resolution))
    grid = torch.tensor(np.c_[xx.ravel(), yy.ravel()], dtype=torch.float32)
    with torch.no_grad():
        preds = model(grid).argmax(dim=1).numpy()
    preds = preds.reshape(xx.shape)
    plt.contourf(xx, yy, preds, alpha=0.3)
    plt.scatter(data[:, 0], data[:, 1], c=labels, s=30, edgecolor='k')
    plt.show()
```
